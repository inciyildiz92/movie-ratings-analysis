```python

import pandas as pd

data = {
    "title": ["Inception", "Interstellar", "The Dark Knight", "Harry Potter 1", "Harry Potter 3", "Titanic", "The Matrix", "Gladiator"],
    "year": [2010, 2014, 2008, 2001, 2004, 1997, 1999, 2000],
    "genre": ["Sci-Fi", "Sci-Fi", "Action", "Fantasy", "Fantasy", "Romance", "Sci-Fi", "Action"],
    "rating": [8.8, 8.6, 9.0, 7.6, 7.9, 7.9, 8.7, 8.5],
    "votes": [2200000, 1800000, 2600000, 800000, 650000, 1200000, 1900000, 1500000]
}

df = pd.DataFrame(data)
df.head()
```

```python

print("Line numbers:", len(df))
print("Avarage rating:", df["rating"].mean())
print("The highest rating:", df["rating"].max())
print("The lowest rating:", df["rating"].min())

top_movie = df.loc[df["rating"].idxmax(), ["title", "rating"]]
print("\nThe highest voted movie:")
print(top_movie)

genre_avg = df.groupby("genre")["rating"].mean().sort_values(ascending=False)
print("\nAvarage rating according to the genres:")
print(genre_avg)
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(7,4))
plt.hist(df["rating"], bins=6, edgecolor="black")
plt.title("Rating Chart")
plt.xlabel("Rating")
plt.ylabel("Movie Numbers")
plt.show()
```

```python

plt.figure(figsize=(7,4))
genre_avg.plot(kind="bar")
plt.title("Avarage Rating According To The Genres")
plt.xlabel("Genre")
plt.ylabel("Avarage Rating")
plt.xticks(rotation=0)
plt.show()
```
